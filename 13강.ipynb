```python
# by DB
# n , k = map(int, input().split())
n, k = 170, 23

result = 0

while True:
    # n이 k로 나누어 떨어지는 수가 될 때까지 빼기
    target = (n//k) * k
    result += (n - target)
    n = target
    # n이 k보다 작을 때
    if n < k:
        break
    # k로 나누기
    result += 1
    n //= k

    
# 마지막으로 남은 수에 대하여 1씩 빼기
result += (n - 1)
print(result)
```

```python
# by DB
data = input()

result = int(data[0])

for i in range(1, len(data)):
    num = int(data[i])
    if num <= 1 or result <= 1:
        result += num
    else:
        result *= num
        
print(result)
```

```python
# 나동빈 코드
n = int(input())
data = list(map(int, input().split()))
data.sort()

result = 0 # 최종 그룹 수
count = 0 # 현재 그룹마다의 사람 수 

for i in data:
    count += 1
    if count >= i:
        result += 1
        count = 0
        
print(result)
```
